```python
import yfinance as yf
```

```python
tickers = ['FA','AMZN','AAPL','NFLX','GOOG']
```

```python
fig, ax = plt.subplots()
for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist = stock.history(start='2021-01-01',end='2021-12-31') 
    ax.plot(hist.index, 100*hist.Close/hist.Close[0], label=ticker)
plt.legend();
```
