```python
import scipy.stats as stats
import matplotlib.pyplot as plt

a = stats.norm(loc=0, scale=1)
x = a.rvs(size=(10000))
plt.hist(x,bins=50)
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=0, scale=1)
x = np.linspace(-3,3)
cdf = a.cdf(x)
plt.plot(x,cdf)
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=0, scale=1)
x = np.linspace(-3,3)
pdf = a.pdf(x)
plt.plot(x,pdf)
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=0, scale=1)
x = np.linspace(0,1)
ppf = a.ppf(x)
plt.plot(x,ppf)
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=0, scale=1)
x_left = 0.025
left_cut = a.ppf(x_left)
x_right = 0.975
right_cut = a.ppf(x_right)
print(left_cut, right_cut)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=185, scale=11)
x_left = 0.1
left_cut = a.ppf(x_left)
print(left_cut)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

a = stats.norm(loc=1, scale=2)
x_left = 0
left_cdf = a.cdf(x_left)
x_right = 1
right_cdf = a.cdf(x_right)
mass = right_cdf - left_cdf
print(mass)
```
