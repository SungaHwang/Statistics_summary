```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = np.random.normal(size=(10_000))
x_fig = np.linspace(-3,3,100)
y_fig = np.exp(-x_fig**2/2) / np.sqrt(2*np.pi)

plt.hist(x, density=True, bins=100);
plt.plot(x_fig, y_fig);
```

```python
x = stats.norm().rvs(size=10_000)
x_fig = np.linspace(-3,3,100)
y_fig = stats.norm().pdf(x_fig)

plt.hist(x, density=True, bins=100)
plt.plot(x_fig, y_fig);
```
