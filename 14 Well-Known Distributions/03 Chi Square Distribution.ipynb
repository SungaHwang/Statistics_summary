```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

$$
\chi^2(n)\sim \sum_{i=1}^nZ_i^2\quad\text{$Z_i$ iid $N(0,1)$}
$$

```python
df = 10
n_samples = 10_000
```

```python
normal_samle = np.random.normal(size=(n_samples, df)) # (10000, 10)
chi2_sample = np.sum(normal_samle**2, axis=1) # (10000,)
```

```python
x_fig = np.linspace(0, 60, 100)
y_fig = stats.chi2(df).pdf(x_fig)
```

```python
plt.hist(chi2_sample, bins=100, density=True)
plt.plot(x_fig, y_fig);
```
