```python
import matplotlib.pyplot as plt; plt.style.use('seaborn')
import numpy as np; np.random.seed(0)
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('data/StudentStatistics.csv', sep=';')
df
```

```python
myCoding = {
    'Fully Disagree': 1, 
    'Disagree': 2, 
    'Neither disagree nor agree': 3, 
    'Agree': 4, 
    'Fully agree': 5
}
```

```python
df['Teach_MotivateRec'] = df['Teach_Motivate'].replace(myCoding)
df
```
