```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(-1,1).reshape((-1,1)) # (50,1)
noise = np.random.normal(0,1,size=x.shape) # (50,1)
noise_level = 0.5
y = (1 + 2*x + 3*x**2 + noise_level * noise).reshape((-1,))
```

```python
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
residue = y - y_pred
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,3))
axes[0].plot(x,y,'o')
axes[0].plot(x,y_pred)
axes[1].plot(x,residue,'o');
```

```python
print(x.shape) # (50,1)
x_2 = np.concatenate([x,x**2],axis=1)
print(x_2.shape) # (50,2)
```

```python
model_2 = LinearRegression()
model_2.fit(x_2,y)
y_pred_2 = model_2.predict(x_2)
residue_2 = y - y_pred_2
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,3))
axes[0].plot(x,y,'o')
axes[0].plot(x,y_pred_2)
axes[1].plot(x,residue_2,'o');
```
